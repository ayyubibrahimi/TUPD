```python
import pandas as pd
```

```python
data = pd.read_csv('../data/input/tupd.csv')

data['cleaned_officer_name'] = data['Officer name & Allegation #'].str.extract(r'(.+?) Allegation')[0].fillna(data['Officer name & Allegation #'])

cleaned_data = data.drop_duplicates(subset='cleaned_officer_name')

# Convert the 'left_date' and 'investigation_complete_date' to datetime format
cleaned_data['left_date'] = pd.to_datetime(cleaned_data['left_date'])
cleaned_data['investigation_complete_date'] = pd.to_datetime(cleaned_data['investigation_complete_date'])

# Calculate the time difference between leaving and investigation completion dates
cleaned_data['time_difference'] = cleaned_data['left_date'] - cleaned_data['investigation_complete_date']

# Define bins for categorizing time differences, including handling very early departures
extended_bins = pd.to_timedelta([-10000, -365, 0, 30, 90, 180, 365, 3650], 'D')  # Extend lower limit to about 27 years
extended_labels = ['Left long before investigation', 'Left before investigation', 'Within 30 days', 'Within 3 months', 'Within 6 months', 'Within a year', 'After a year']

# Categorize each officer into time buckets
cleaned_data['extended_time_bucket'] = pd.cut(cleaned_data['time_difference'], bins=extended_bins, labels=extended_labels, right=False)

extended_officer_time_buckets = cleaned_data[['cleaned_officer_name', 'extended_time_bucket']]
extended_officer_time_buckets
```
